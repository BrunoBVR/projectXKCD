Loop through all strips till the response from url is 404.

```python
url = "https://xkcd.com/1/info.0.json"

try:
    # store the response of URL
    response = urlopen(url)

    # storing the JSON response 
    # from url in data
    data_json = json.loads(response.read())
    
    # print the json response
    print(data_json)
except:
    print("Last strip!")
```

# Getting all strips

```python
final_strip = False
counter = 1
data_list = []

# import urllib library
from urllib.request import urlopen
  
# import json and pandas
import json
import pandas as pd

while not final_strip:

    print(f"On page: {counter}")
    try:
        # store the URL in url as parameter for urlopen
        url = f"https://xkcd.com/{counter}/info.0.json"

        # store the response of URL
        response = urlopen(url)
        
        # storing the JSON response from url in data
        data_json = json.loads(response.read())
        
        # append the json response to data_list
        data_list.append(data_json)

        # go to next page
        counter += 1
        if counter == 404:
            print("LOL, strip 404 is empty!")
            # Randall Munroe actually skipped strip 404 as a joke, to make it an actual 404 error!
            counter += 1
    except:
        print("Reached final strip!")
        final_strip = True
```

```python
len(data_list)
```

```python
# Save data_list to pickle file
import pickle

with open('data_list.pickle', 'wb') as handle:
    pickle.dump(data_list, handle)
```

```python
# Getting data into pandas df
xkcd_df = pd.DataFrame(data_list)
```
